```python
from sentence_transformers import SentenceTransformer

# Load model
model = SentenceTransformer("intfloat/multilingual-e5-base")

# 1. Siapkan 5 Kalimat (Chunks) dan 1 Pertanyaan
sentences = [
    "Cuaca di Jakarta hari ini sangat cerah dan panas.",    # Kalimat 0 (Relevan)
    "Harga saham teknologi sedang naik drastis minggu ini.", # Kalimat 1
    "Kucing saya suka tidur di atas sofa warna biru.",       # Kalimat 2
    "Jangan lupa bawa payung karena nanti sore akan hujan.", # Kalimat 3 (Agak relevan)
    "Tim sepak bola itu memenangkan piala dunia tahun lalu." # Kalimat 4
]

query = "Bagaimana perkiraan cuaca hari ini?"

# 2. Preprocessing: Tambahkan prefix khusus E5
# E5 membutuhkan 'passage: ' untuk dokumen dan 'query: ' untuk pertanyaan
input_docs = ["passage: " + s for s in sentences]
input_query = "query: " + query

# 3. Encode (ubah jadi vektor)
# normalize_embeddings=True penting untuk Cosine Similarity yang akurat
doc_embeddings = model.encode(input_docs, normalize_embeddings=True)
query_embedding = model.encode(input_query, normalize_embeddings=True)

# 4. Hitung Kemiripan (Similarity)
# Ini akan membandingkan 1 query vector vs 5 doc vectors
similarities = model.similarity(query_embedding, doc_embeddings)

# 5. Tampilkan Hasil
print(f"Pertanyaan: '{query}'\n")
print("Skor Kemiripan (Cosine Similarity):")
print("-" * 50)

# similarities[0] karena kita hanya punya 1 query
for i, score in enumerate(similarities[0]):
    print(f"Score: {score:.4f} | Kalimat: {sentences[i]}")

# (Opsional) Mengurutkan dari yang paling mirip (Ranking)
print("\n--- Ranking (Paling Relevan) ---")
# Menggabungkan score dan kalimat, lalu sort descending
ranked_results = sorted(zip(similarities[0], sentences), key=lambda x: x[0], reverse=True)

for score, sentence in ranked_results:
    print(f"Score: {score:.4f} -> {sentence}")
```
